```python
"""1. Write a function that inputs a number and prints the multiplication table of that number"""


def input_number():
    number = int(input("Enter a number that you want the table for! \n******************************************* \n\n"))
    multiplication(number)

def multiplication(number):
    try:
        if(number == 0):
            raise ValueError("\n\nPlease enter a non-zero number. Both negative and postive numbers are accepted")
        else:    
            print("\n\nThe multiplication table for {} till 15 is: \n******************************************* \n\n".format(number))
            for i in range(1, 16):
                result = 0
                result = result + number*i
                print("{} x {} = {}".format(number,i,result)) 
                
    except ValueError as exception:
        print(exception)
        input_number()

input_number()
```

```python
"""2. Write a program to print twin primes less than 1000. If two consecutive odd numbers are both prime then they are known as twin primes"""

import numpy as np

limit = 1000
lst = [] 

def check_prime(number):
    count = 0
    for i in range(2, number):
        if(number % i == 0):
            count = count + 1
    if(count == 0):
        return True
    else:
        return False
    
print("The twin primes less than {} are : \n\n".format(limit))
    
for i in range(3, limit, 2):
    
    if(check_prime(i) & check_prime(i+2)):
        print("{} and {}".format(i, i+2))
        
```

```python
"""4. Write a program to implement these formulae of permutations and combinations. Number of permutations of n objects taken r at a time: p(n, r) = n! / (n-r)!. 
      Number of combinations of n objects taken r at a time is: c(n, r) = n! / (r!*(n-r)!) = p(n,r) / r!"""


def input_number():
    n = int(input("Enter the total number of objects, 'n':\n"))
    r = int(input("Enter the total number of objects to be taken at each time, 'r':\n"))
    if(n > r):
        permutation(n, r)
        combination(n, r)
    else:
        print("\n\nInvalid Entry! Please note: The value of n has to be more than r. \nPlease try again!")
        input_number()
            
def factorial(number):
    result = 1
    for i in range(1, number+1):
        result = result * i
    return result

def permutation(n, r):
        permutation = factorial(n) / factorial(n-r)
        perm = int(permutation)
        print("\nNumber of permutations of {} objects taken {} at a time is:\n{}".format(n, r, perm))
        
def combination(n, r):
        combination = factorial(n) / (factorial(r) * factorial(n-r))
        comb = int(combination)
        print("\nNumber of combinations of {} objects taken {} at a time is:\n{}".format(n, r, comb))
    
input_number()
```

```python
"""5. Write a function that converts a decimal number to binary number"""


number = int(input("Please enter a decimal number: \n\n"))
lst = []
def decimal_to_binary(number):
    while(number != 0):
        rem = number % 2
        lst.append(rem)
        number = number // 2
    lst.reverse()
        
decimal_to_binary(number)
string = ''
for i in lst:
    string += str(i)
     
print("\n\nThe binary representation of number {} is: \n\n{}".format(number,string))        
```

```python
"""6. Write a function cubesum() that accepts an integer and returns the sum of the cubes of individual digits of that number. 
      Use this function to make functions PrintArmstrong() and isArmstrong() to print Armstrong numbers and to find whether is an Armstrong number."""

number = int(input("Enter any number which is an integer: \n"))
limit = 1000

def cubesum(number):
    sum_of_digits = 0
    while(number != 0):
        temp = number % 10
        sum_of_digits = sum_of_digits + temp**3
        number = number // 10
    return sum_of_digits

def isArmstrong(number):
    sum_of_digits = cubesum(number)
    if(sum_of_digits == number):
        print("{} is an armstrong number!\n".format(number))
    else:
        print("{} is not an armstrong number!\n".format(number))
        
def PrintArmstrong(limit):
    lst = []
    for i in range(1, limit+1):
        sum_of_digits = cubesum(i)
        if(sum_of_digits == i):
            lst.append(i)
    print("All armstrong numbers within {} are:\n{}".format(limit, lst))
    
sum_of_digits = cubesum(number)
print("\nThe sum of the cubes of individual digits of number {} is:\n{}\n".format(number, sum_of_digits))
isArmstrong(number)
PrintArmstrong(limit)
```

```python
"""7. Write a function prodDigits() that inputs a number and returns the product of digits of that number."""

def prodDigits():
    list_of_digits = []
    original_input_number = int(input("Please enter a number you want to find the product of digits for!\n"))
    number = original_input_number
    product_of_digits = 1
    while(number != 0):
        temp = number % 10
        list_of_digits.append(temp)
        number = number // 10
    
    for i in list_of_digits:
        product_of_digits = product_of_digits * i

    print("\nThe product of digits of {} is: \n{}".format(original_input_number, product_of_digits))
    
prodDigits()
```

```python
"""9. Write a function sumPdivisors() that finds the sum of proper divisors of a number. Proper divisors of a number are those numbers by which the number is divisible, 
except the number itself. For example proper divisors of 36 are 1, 2, 3, 4, 6, 9, 12, 18"""


number = int(input("Enter a number you want to find the proper divisors for:\n"))

def sumPdivisors(number):
    list_divisors = []
    for i in range(1, number):
        if(number % i == 0):
            list_divisors.append(i)
    print("\nThe proper divisors of {} are:\n{}".format(number, list_divisors))
    print("\nThe sum of proper divisors of {} is:\n{}".format(number, sum(list_divisors)))
    
sumPdivisors(number)
```

```python
"""10. A number is called perfect if the sum of proper divisors of that number is equal to the
       number. For example 28 is perfect number, since 1+2+4+7+14=28. Write a program to
       print all the perfect numbers in a given range"""

limit = int(input("Enter a range to find all perfect numbers within the given range:\n"))
def check_perfect(number):
    list_of_divisors = []
    for i in range(1, number):
        if(number % i == 0):
            list_of_divisors.append(i)
    if(sum(list_of_divisors) == number):
        return True
    else:
        return False
        
def print_perfect(limit):
    all_perfect_numbers = []
    for numbers in range(1, limit+1):
        if check_perfect(numbers):
            all_perfect_numbers.append(numbers)
    print("The perfect numbers within {} are:\n{}".format(limit, all_perfect_numbers))
    
        
print_perfect(limit)
```

```python
"""12. Write a program which can filter odd numbers in a list by using filter function"""

import numpy as np

array_list = np.random.randint(1, 1000, 100)
lst = list(array_list)

def filter_odd_numbers(lst):
    odd_list = list(filter(lambda x : (x%2 == 1),lst))
    print(odd_list)
    
filter_odd_numbers(lst)
```
